```python
def NandM_population(N_0, M_0, K_N, K_M, S, d_N, d_M, time):
    # Exponential growth model
    # Save N_values into a list
    N_values = [N_0]
    M_values = [M_0]
    # Initialize list
    # Set current population
    current_N = N_0
    current_M = M_0
    # Loop through time steps
    for i in range(time):
        if i == 4:
            current_M = current_M + 100
        # Update population
        current_N = current_N + S * current_N * (1 - (current_N / K_N) - (current_M / K_M)) - d_N * current_N
        current_M = current_M + S * current_M * (1 - (current_M / K_M) - (current_N / K_N)) - d_M * current_M
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        current_M = round(current_M, 3)
        # Append to list
        N_values.append(current_N)
        M_values.append(current_M)
    return N_values, M_values
# Parameters
S = 1 # intrinsic growth rate
d_N = 0.4 # death rate of species N
d_M = 0.6 # death rate of species M
K_N = 400 # carrying capacity of species N
K_M = 1600 # carrying capacity of species M
N_0 = 12 # initial population of species N
M_0 = 5 # initial population of species M
time = 38 # time in years
```

```python
# Run the function with time = 38, K_N = 400, K_M = 1600, S = 1, d_N = 0.4, d_M = 0.6, N_0 = 12, M_0 = 5
#NandM_population(N_0, M_0, K_N, K_M, S, d_N, d_M, time)
resultN,resultM = NandM_population(12, 5, 400, 1600, 1, 0.4, 0.6, 38)
print(resultN)
```

```python
import matplotlib.pyplot as plt
# Calculate population results
resultN, resultM = NandM_population(N_0, M_0, K_N, K_M, S, d_N, d_M, time)
# Plot the results with years on the x-axis and population on the y-axis
years = list(range(len(resultN)))
plt.plot(years, resultN, marker='o', label='N Population')
plt.plot(years, resultM, marker='s', label='M Population')
plt.xlabel('Years')
plt.ylabel('Population Size')
plt.title('Population Growth of N and M Over Time')
plt.legend()
plt.grid()
plt.show()
```
