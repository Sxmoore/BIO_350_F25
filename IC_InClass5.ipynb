```python
# Create a real-world example of an age-structured model with 5 populations with values such as F (number of offspring per individual) and S (survival rate from one individual to the next).
import numpy as np
import matplotlib.pyplot as plt
# Define parameters for the age-structured model
age_classes = 5
# Age class parameters
#Age class 1: 0-1 years
#Age class 2: 1-2 years
#Age class 3: 2-3 years
#Age class 4: 3-4 years
#Age class 5: 4+ years
# Fertility rates (F) and survival rates (S)
# F: number of offspring per individual in each age class
F = np.array([0, 0, 1.2, 1.5, 1.0])  # Fertility rates for each age class
S = np.array([0.5, 0.7, 0.8, 0.7, 0.4])    # Survival rates between age classes
#Leslie matrix construction
L = np.zeros((age_classes, age_classes))
# Fill in the first row with fertility rates
L[0, :] = F
# Fill in the sub-diagonal with survival rates
for i in range(1, age_classes):
    L[i, i-1] = S[i-1]
# Initial population distribution across age classes
initial_population = np.array([100, 80, 60, 40, 20])
# Number of time steps to simulate
time_steps = 50
# Store population over time
population_over_time = np.zeros((time_steps, age_classes))
population_over_time[0, :] = initial_population
# Simulate population dynamics over time
for t in range(1, time_steps):
    population_over_time[t, :] = L @ population_over_time[t-1, :]
# Plot the results
plt.figure(figsize=(10, 6)) 
for i in range(age_classes):
    plt.plot(population_over_time[:, i], label=f'Age class {i+1}')
plt.title('Age-Structured Population Dynamics')
plt.xlabel('Years')
plt.ylabel('Population Size')
plt.legend()
plt.grid()
plt.show()


```

# What was the real world example?
- My real-world example was an age-structured model of a deer population divided into 5 age classes: fawns, yearlings, young adults, adults, older adults.

- Based on my results, there is a greater number of fawns compared to all other age classes existing within the population.

- The transient stage is between years 0 through 10. During this period, each age class experienced a rise and fall in the number of individuals. This inindicated by the fluctuating amplitude and shorter fequency of the lines. Eventually, the lines level out and slowly begin to increase over the next 40 years. This is the stable stage of the population dynamic.
