Plot the data showing the frequency of your data, with the probability density on the y axis

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import nbinom

# Parameters
mu = 4       # mean encounters
k = 2        # dispersion parameter
n = 20       # number of surveys

# Convert to standard nbinom parameters (number of failures k, probability p)
p = k / (k + mu)

# Simulate data
np.random.seed(42)
encounters = nbinom.rvs(k, p, size=n)

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(encounters, bins=range(0, max(encounters)+2), kde=False, color="forestgreen")
plt.title("Simulated Moose Encounter Counts (Negative Binomial)", fontsize=14)
plt.xlabel("Number of Moose Encounters per Survey")
plt.ylabel("Frequency")
plt.grid(True, alpha=0.3)
plt.show()

# Summary statistics
mean_obs = np.mean(encounters)
var_obs = np.var(encounters, ddof=1)

print(f"Observed Mean = {mean_obs:.2f}")
print(f"Observed Variance = {var_obs:.2f}")
```
