```python
#Create a function that takes initial S, I, R, beta, death rate, and time as inputs
def SIR_model(S0, I0, R0, beta, d_rate, time):
    #Import the variables given to the function
    S_values = [S0]
    I_values = [I0]
    R_values = [R0]
    Current_S = S0
    Current_I = I0
    Current_R = R0
    #For loop to go through time period
    for time in range(1, time + 1):
        new_infections = beta * Current_S * Current_I
        new_recoveries = d_rate * Current_I
        Current_S -= new_infections
        Current_I += new_infections - new_recoveries
        Current_R += new_recoveries
        S_values.append(Current_S)
        I_values.append(Current_I)
        R_values.append(Current_R)
    #return 3 variables seperated by comma
    return S_values, I_values, R_values
#Initial parameters
S0 = 999
I0 = 1
R0 = 0
beta = 0.001
d_rate = 0.25
time = 30
#Call the SIR_ and give it S0, I0, R0, beta, d_rate, time and save them into 3 variables (S_list, I_list, R_list)
S_list, I_list, R_list = SIR_model(S0, I0, R0, beta, d_rate, time)
```

```python
#Plot the results
plt.plot(S_list, label='Susceptible')
plt.plot(I_list, label='Infected')
plt.plot(R_list, label='Removed')
plt.xlabel('Days')
plt.ylabel('Number of Individuals')
plt.grid()
plt.legend()
plt.title('SIR Model Simulation')
plt.show()
```

# Beta variable - What happens when it is changed up or down? What are the reasons it may change?
The increase and decrease in the beta variable changes the infection and removal rates. When beta is increased (>), there is a greater chance for individuals within a population to be infected and removed. Likewise, when it is decreased (<), it lowers (<) those likelyhoods. Some potential factors that may cause the beta to rise or fall is the total habitat space and the distance between individuals.
