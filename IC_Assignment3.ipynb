- Calculating how long it takes a plot of land to be covered in plants.

```python
import matplotlib.pyplot as plt
```

```python
def plot_growth(N_0, b, d, time):
    #Logistic growth model
    #Save N_values into a list
    N_values = [N_0]
    #Save initial population
    Current_N = N_0
    #Loop through time steps
    for _ in range(time):
        #Update population
        Current_N = Current_N + (b * Current_N) - (d * Current_N)
        #Round to 3 decimal places
        Current_N = round(Current_N, 3)
        #Append to list
        N_values.append(Current_N)
    return N_values
```

```python
# Run the function with time = 20, b = 1, d = 0.4, with an initial population of 12
result = plot_growth(12, 1, 0.4, 20)
print(result)
```

```python
# Plot the results with time on the x-axis and population on the y-axis
time = list(range(21))  # 0 to 20 time
# create the plot
plt.plot(time, result, marker='o')
# Add labels and title
plt.xlabel('Years')
plt.ylabel('Plant Population (proportion of pond covered)')
plt.title('Plant Population Growth Over Time')
# Add grid
plt.grid()
# Show the plot
plt.show()
```

```python
#Create a self-shading formula
# Logistic growth with self-shading effect
def logistic_growth(N, r, K, time):
    N_values = [N]
    Current_N = N
    for _ in range(time):
        Current_N = Current_N + r * Current_N * (1 - Current_N / K)
        Current_N = round(Current_N, 3)
        N_values.append(Current_N)
    return N_values
#Parameters
r = 1  # intrinsic growth rate
d = 0.4  # death rate
K = 400   # carrying capacity
N_0 = 12  # initial population
time = 20  # time steps
# Run the logistic growth function
result = logistic_growth(N_0, r, K, time)
print(result)
```

```python
# Plot the results with time on the x-axis and population on the y-axis
time = list(range(21))  # 0 to 20 time
# create the plot  
plt.plot(time, result, marker='o')
# Add labels and title
plt.xlabel('Years')
plt.ylabel('Plant Population (proportion of pond covered)')
plt.title('Plant Population Growth Over Time with Self-Shading Effect')
# Add grid
plt.grid()
# Show the plot
plt.show()
```

```python
# Logistic growth with self-shading effect accounting for intraspecific compatition
def logistic_growth(N, s, d, K, time):
    N_values = [N]
    Current_N = N
    for _ in range(time):
        Next_N = Current_N + s * Current_N * (1 - Current_N / K) - d * Current_N
        Next_N = round(Next_N, 1)
        N_values.append(Next_N)
        Current_N = Next_N  # <-- update for next iteration
    return N_values

# Parameters
s = 1  # intrinsic growth rate
d = 0.4  # death rate
K = 400   # carrying capacity
N_0 = 12  # initial population
time = 20  # time steps

# Run the logistic growth function
result = logistic_growth(N_0, s, d, K, time)
print(result)
```

```python
# Plot the results with time on the x-axis and population on the y-axis
time = list(range(21))  # 0 to 20 time
# create the plot  
plt.plot(time, result, marker='o')
# Add labels and title
plt.xlabel('Years')
plt.ylabel('Plant Population (proportion of pond covered)')
plt.title('Plant Population Growth Over Time with Self-Shading Effect')
# Add grid
plt.grid()
# Show the plot
plt.show()
```
